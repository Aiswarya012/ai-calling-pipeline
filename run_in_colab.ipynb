```python
# Make a Hindi test input with Kokoro (so we don't need to upload a WAV)
from kokoro import KPipeline
import soundfile as sf, numpy as np
kp = KPipeline(lang_code='h')
text = "नमस्ते, मेरा नाम राहुल है, मुझे कल शाम चार बजे टेस्ट ड्राइव के लिए अपॉइंटमेंट चाहिए।"
sf.write('input_hi.wav', np.concatenate([a for _, _, a in kp(text, voice='hf_alpha')]), 24000)
print('wrote input_hi.wav')
```
